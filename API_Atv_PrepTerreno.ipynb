```python
# ------------------------------
# Etapa 1: Preparar o Terreno
# ------------------------------

import requests
import pandas as pd

# localização (exemplo: Ciudad del Este)
lat = -25.5167
lon = -54.6167

# quantidade de dias passados
quantos_dias = 7

# endpoint da API
url = 'https://api.open-meteo.com/v1/forecast'

# parâmetros utilizados
parametros = {
    'latitude': lat,
    'longitude': lon,
    'daily': ['temperature_2m_max', 'precipitation_sum'],
    'timezone': 'auto',
    'past_days': quantos_dias
}

# -----------------------------------
# Etapa 2: Coleta de Dados e Verificação
# -----------------------------------

# fazer requisição
response = requests.get(url, params=parametros)

# verificar se deu certo
# The original code had 'resposta' which was not defined, changing to 'response'
if response.status_code == 200:
    print("Requisição feita com sucesso!")

    # transformar em JSON
    dados = response.json()

    # pegar dados diários
    dados_diarios = dados['daily']

    # criar DataFrame
    df_dados_clima = pd.DataFrame(dados_diarios)

    # exibir
    print(df_dados_clima)

else:
    print(f"Erro na requisição: {response.status_code}")

# -----------------------------------
# Etapa 3: Salvar os Dados (CSV)
# -----------------------------------

df_dados_clima.to_csv('relatorio_prof_luiz.csv', index=False)

print("Arquivo salvo com sucesso!")

# -----------------------------------
# EXTRA: Converter para Fahrenheit
# -----------------------------------

df_dados_clima['temperature_2m_max_F'] = (df_dados_clima['temperature_2m_max'] * 1.8) + 32

print("Dados com temperatura em Fahrenheit:")
print(df_dados_clima)
```
